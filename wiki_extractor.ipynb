```python
import wikipedia as wk
!pip install lxml
import json
import requests
from bs4 import BeautifulSoup
```

```python
def wik_Search(search_word, n, file_name):
  x = wk.search(search_word,n)
  l = list()
  for i in x:
    url = wk.page(i).url
    page = requests.get(url)
    soup = BeautifulSoup(page.content,"lxml")
    y = soup.find_all('p')
    print(y[1].text)
    l.append({"url":url,"paragraph":y[1].text})
  with open(file_name+".json", "w") as outfile:
    json.dump(l, outfile)
  return(l)
    
```

```python
l = wik_Search("Bills", 10,"WIKI_EXTRACTED")
```

```python
import urllib
# from urllib.request import Request, urlopen
from PyPDF2 import PdfFileWriter, PdfFileReader
from io import  BytesIO
def pdf_make(url, file_name):
  url = url
  writer = PdfFileWriter()

  remoteFile = urllib.request.urlopen(urllib.request.Request(url)).read()
  memoryFile =  BytesIO(remoteFile)
  pdfFile = PdfFileReader(memoryFile)

  for pageNum in range(pdfFile.getNumPages()):
          currentPage = pdfFile.getPage(pageNum)
          #currentPage.mergePage(watermark.getPage(0))
          writer.addPage(currentPage)


  outputStream = open(file_name,"wb")
  writer.write(outputStream)
  outputStream.close()
```
